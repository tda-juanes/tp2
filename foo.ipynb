```python
import random
import time
from matplotlib import pyplot as plt
from main import maximizar_ganancias

```

funcion para generar listas random

```python
# la energ√≠a siempre va decreciendo
def generar_energias_random(cantidad_dias):
    energias_rand = [random.randint(1, 100) for _ in range(cantidad_dias)]
    energias_rand.sort(reverse=True)
    return energias_rand

def generar_entrenamiento_random(cantidad_dias):
    energias = generar_energias_random(cantidad_dias)
    entrenamientos = [random.randint(1, 100) for _ in range(cantidad_dias)]
    return energias, entrenamientos
```

```python
inicio = 100
limite = 3000
step = 1
conjuntos_entrenamientos = []

# genero los conjuntos de entrenamientos
for i in range(inicio, limite, step):
    conjuntos_entrenamientos.append(generar_entrenamiento_random(i))

# mido el tiempo de ejecuci√≥n de cada conjunto
tiempos = [] # (cantidad_dias, tiempo)
for i in range(len(conjuntos_entrenamientos)):
    timer_start = time.perf_counter()
    maximizar_ganancias(conjuntos_entrenamientos[i][0], conjuntos_entrenamientos[i][1])
    timer_end = time.perf_counter()
    tiempos.append((i+1, timer_end - timer_start))

print(tiempos)
```

```python
plt.plot([t[0] for t in tiempos], [t[1] for t in tiempos])
plt.xlabel('cantidad de d√≠as')
plt.ylabel('tiempo de ejecuci√≥n (s)')


plt.show()
```
